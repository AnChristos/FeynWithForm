```python
%%pyForm PionDecay

* Process: pion -> fermion anti-neutrino

#-
* Above suppresses extra output
Off Statistics;
Off FinalStats;
#include SquareAmplitude.h

Symbols  ml, mpi;
Vector pPion;

* This should be multiplied by a constant which I ommit here.
Local M =  UB(i1, p3, ml) *  g(i1, i2, k7) * g(i2, i3, mu1) * V(i3, p4, 0)  * pPion(mu1);

#call squareamplitude(M, Msq)
.sort

* --- Kinematics ---
id p3.p3 = ml^2;
id p4.p4 = 0;
id pPion.pPion = mpi^2;
id p3.p4 = (mpi^2 - ml^2)/2;
id pPion.p3 = (mpi^2 + ml^2)/2;
id pPion.p4 = (mpi^2 - ml^2)/2;
.sort

* Save to file 
Format C;
#write <PionDecay.txt> "%e;", Msq;
.sort
* Print 
Format;
Print Msq;


.end
```

```python
import numpy as np
import sympy as sp
from Kinematics import dGamma_dOmega


# Capture Raw FORM ouput
form_expr = capture_physics_expr("scripts/PionDecay.txt")
# Convert FORM to sympy
mpi, ml = sp.symbols("mpi ml")
Msq = sp.simplify(form_expr)
print(f"<|M^2|> =  {Msq}")

dgamma = dGamma_dOmega(Msq, mpi, ml, 0)
print(f"dgamma =  {dgamma}\n")

me, mmu = sp.symbols("me mmu", positive=True)
dgamma_e = dgamma.subs(ml, me)
dgamma_mu = dgamma.subs(ml, mmu)
ratio_expr = sp.simplify(dgamma_e / dgamma_mu)
print(f"Ratio: {ratio_expr}\n")

f_ratio = sp.lambdify((me, mmu, mpi), ratio_expr, "numpy")

print(f"Ratio electron to muon: {f_ratio(0.000511,0.10566,0.13957):.2e}")
```
