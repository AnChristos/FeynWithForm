```python
import numpy as np
import sympy as sp
from Kinematics import dGamma_dOmega


# Capture Raw FORM ouput
form_expr = capture_physics_expr("PionDecay.txt")
# Convert FORM to sympy
mpi, ml = sp.symbols("mpi ml")
Msq = sp.simplify(form_expr)
print(f"Msq :")
sp.pprint(Msq, num_columns=80)
print("\n")

dgamma = dGamma_dOmega(Msq, mpi, ml, 0)
print(f"dgamma :")
sp.pprint(dgamma, num_columns=80)
print("\n")

me, mmu = sp.symbols("me mmu", positive=True)
dgamma_e = dgamma.subs(ml, me)
dgamma_mu = dgamma.subs(ml, mmu)
ratio_expr = sp.simplify(dgamma_e / dgamma_mu)
print(f"ratio :")
sp.pprint(ratio_expr, num_columns=80)
print("\n")

f_ratio = sp.lambdify((me, mmu, mpi), ratio_expr, "numpy")

print(f"Ratio electron to muon: {f_ratio(0.000511,0.10566,0.13957):.2e}")
```
