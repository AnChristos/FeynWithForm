## One-loop QED vacuum polarization (electron bubble)

We compute the one-loop correction to the photon propagator coming from an electron loop. 
The object of interest is the tensor integral

## Dirac trace

After evaluating the trace, the integrand becomes

$$ \Pi^{\mu\nu}(q) \sim \int \frac{d^d k}{(2\pi)^d}
\frac{
-4 k^\mu (k-q)^\nu
-4 k^\nu (k-q)^\mu
-4 g^{\mu\nu} m^2
+4 g^{\mu\nu} k\cdot (k-q)
}{
(k^2 - m^2)((k-q)^2 - m^2)
} $$

---

## Propagators and rewriting

We define propagators:

- $ D_1 = k^2 - m^2 $
- $ D_2 = (k - q)^2 - m^2 $

and rewrite all scalar products in terms of them:

$$ k \cdot q = \frac{D_1 - D_2 + q^2}{2} $$

This step rewrites the numerator entirely in terms of:
- $k^\mu k^\nu$
- $k^\mu$
- scalar quantities

so that standard integral reductions can be applied.

---

## Tensor integral reduction

We reduce all loop integrals to standard scalar functions.

### Rank-2 tensor integral

$$ \int \frac{k^\mu k^\nu}{D_1 D_2} = g^{\mu\nu} B_{00}(q^2,m,m) + q^\mu q^\nu B_{11}(q^2,m,m) $$

### Rank-1 vector integral

$$ \int \frac{k^\mu}{D_1 D_2}= q^\mu B_1(q^2,m,m) $$

### Scalar integral
$$ \int \frac{1}{D_1 D_2} = B_0(q^2,m,m) $$

---

## Tadpoles

Single-propagator integrals give tadpoles:

### Scalar tadpole
$$ \int \frac{1}{D_1} = A_0(m), \quad
\int \frac{1}{D_2} = A_0(m) $$

### Vector tadpole
- Symmetric case:
  $$ \int \frac{k^\mu}{D_1} = 0 $$

- Shifted case:
To evaluate the rank-1 integral over $D_2$, we perform the shift $k \to k + q$:

$$\int \frac{d^d k}{(2\pi)^d} \frac{k^\mu}{(k-q)^2 - m^2} = \int \frac{d^d k}{(2\pi)^d} \frac{(k+q)^\mu}{k^2 - m^2}= \underbrace{\int \frac{d^d k}{(2\pi)^d} \frac{k^\mu}{k^2 - m^2}}_{0 \text{ (Symmetry)}} + q^\mu \underbrace{\int \frac{d^d k}{(2\pi)^d} \frac{1}{k^2 - m^2}}_{A_0(m)}$$

Thus, or the shifted propagator is:
$$\int \frac{k^\mu}{D_2} = q^\mu A_0(m)$$

---

## Result
After all replacements, the loop becomes
$$ \Pi^{\mu\nu}(q) = g^{\mu\nu} \left( 4 A_0(m) - 2 q^2 B_0(q^2,m,m) - 8 B_{00}(q^2,m,m) \right) + q^\mu q^\nu \left( 8 B_1(q^2,m,m) - 8 B_{11}(q^2,m,m) \right) $$

---

## Ward Identity

If we write the amplitude for a process involving a photon as
$$
\mathcal{M} = \epsilon^\mu(q)\,\mathcal{M}_\mu,
$$
where $\epsilon^\mu$ is the polarization vector, the Ward Identity states:
$$
q^\mu \mathcal{M}_\mu = 0.
$$

This means that replacing $\epsilon^\mu$ with the photon momentum $q^\mu$ gives zero.

For the vacuum polarization tensor:
$$
q^\mu \Pi_{\mu\nu}(q) = 0.
$$

This provides a strong consistency check on the calculation: the result must be transverse.  
In practice, this allows us to focus directly on extracting the transverse structure, since any longitudinal contribution must cancel.

---

## Projector

The projector
$$
P^{\mu\nu} = g^{\mu\nu} - \frac{q^\mu q^\nu}{q^2}
$$
isolates the transverse part of the tensor by removing anything proportional to $q^\mu$, using
$$
q_\mu P^{\mu\nu} = 0.
$$

Since the Ward Identity guarantees that longitudinal pieces are unphysical, applying this projector does not discard real information.

---

## Scalar reduction 

The tensor structure is reduced to scalar integrals. 
The standard relations between Passarino–Veltman functions are used:
$$
B_1(q^2,m,m) = -\frac{1}{2} B_0(q^2,m,m),
$$
$$
B_{11}(q^2,m,m) = \frac{A_0(m) - 2 B_{00}(q^2,m,m)}{2 q^2},
$$
$$
B_{00}(q^2,m,m) =
\frac{A_0(m) + \left(2m^2 - \frac{q^2}{2}\right) B_0(q^2,m,m)}{2(d-1)}.
$$

The scalar integrals evaluate to
$$
A_0(m) = \frac{m^2}{16\pi^2}
\left(\frac{1}{\epsilon} - \log\frac{m^2}{\mu^2} + 1 \right),
$$
$$
B_0(q^2,m,m) = \frac{1}{16\pi^2}
\left(\frac{1}{\epsilon} - \log\frac{q^2}{\mu^2} + 2 \right).
$$

Working in dimensional regularization with
$$
d = 4 - 2\epsilon,
$$
we also use the expansions
$$
\frac{1}{2(d-1)} = \frac{1}{6}\left(1 + \frac{2}{3}\epsilon \right),
$$
$$
\frac{1}{[2(d-1)]^2} = \frac{1}{36}\left(1 + \frac{4}{3}\epsilon \right).
$$



```python
%%pyForm BubbleLoop

* Process: BubbleLoop

#-
* Above suppresses extra output
Off Statistics;
Off FinalStats;
Symbol d;
Dimension d;
#include FeynHelpers.h

Vectors k1,k2;
Symbols e, Mmuon, Melec, y, D1, D2shift;
Symbols bubbleProp, LoopFactor,ep, Scale, pi, Q2, alpha;
CF  A, A0, B, B0, B00, B1, B11, log;


Local Bubble = (-e^2) 
             * gamma(i1, i2, mu1) * fprop(i2, i3, k1, Melec) 
             * gamma(i3, i4, mu2) * fprop(i4, i1, k2, Melec);
#call Propagators()
#call doTrace(1,0)
.sort;
id k2 = k1 - q;
.sort

* Replace propagators with D1, D2shift 
id prop( - Melec^2 + k1.k1) = 1/D1 ;
id prop( - Melec^2 + k2.k2) = 1/D2shift;
.sort

Local Numerator = Bubble * D1 * D2shift  ;
.sort
* Drop the bubble for now
Drop Bubble;
.sort;

* shift/replace loop momentum
id k1.k1 = D1 + Melec^2;
id k1.q  = (D1 - D2shift + q.q)/2;
.sort;

* Prepeare for Ward identity
Local WardNum = q(mu1) * Numerator;
.sort
* Re-apply the scalar product identity to the new q.k1 terms
id k1.q = (D1 - D2shift + q.q)/2;
.sort

* Put the propagators back to turn the numerator into an integral
Local WardInt= WardNum * D1^-1 * D2shift^-1;
Local BubbleInt=  Numerator * D1^-1 * D2shift^-1;
.sort
Drop Numerator, WardNum;
.sort

* Apply the cancellations
id D1 * D1^-1 * D2shift^-1 = D2shift^-1;
id D2shift * D1^-1 * D2shift^-1 = D1^-1;
.sort

* Now do Integral replacement
* Tensor Integral (Rank 2)
id k1(mu1?)*k1(mu2?)*D1^-1*D2shift^-1 = d_(mu1,mu2)*B00(q.q,Melec,Melec) 
                                 + q(mu1)*q(mu2)*B11(q.q,Melec,Melec);
.sort
* Vector Integral (Rank 1)
id k1(mu1?)*D1^-1*D2shift^-1 = q(mu1)*B1(q.q,Melec,Melec);
.sort
* Scalar Integral (Rank 0)
id D1^-1*D2shift^-1 = B0(q.q,Melec,Melec);
.sort 
* Vector Tadpole
* The Symmetric Propagator (D1)
id k1(mu1?)*D1^-1 = 0;
* The Shifted Propagator (D2shift)
id k1(mu1?)*D2shift^-1 = q(mu1)*A0(Melec);
.sort 
* Leftover Tadpoles (Rank 0)
id D1^-1 = A0(Melec);
id D2shift^-1 = A0(Melec);
.sort
#message ">>>"
Print +s WardInt ;
.sort;
Skip WardInt;
.sort

* Ward identify should be zero 

* Now we can proceed with PV replacements
id B1(q.q,Melec,Melec) = -1/2 * B0(q.q,Melec,Melec);
id B11(q.q,Melec,Melec) = (A0(Melec) - 2*B00(q.q,Melec,Melec)) / (2*Q2);
id B00(q.q,Melec,Melec) = (A0(Melec) + (2*Melec^2 - Q2/2)*B0(q.q,Melec,Melec)) / (2*(d-1));
.sort 
Local PiRaw = BubbleInt / Q2;
.sort

id B0(q.q, Melec, Melec) = B0(Q2, Melec, Melec);
id A0(Melec) = LoopFactor * Melec^2 * (ep^-1 - log(Melec^2/Scale^2) + 1);
id B0(Q2, Melec, Melec) = LoopFactor * (ep^-1 - log(Q2/Scale^2) + 2);
id LoopFactor = 1/(16*pi^2);
.sort

id d = 4 - 2*ep;
id ( - 2 + 2*d)^-1 = 1/6 * (1 + (2/3)*ep);
id ( - 2 + 2*d)^-2 = 1/36 * (1 + (4/3)*ep);
.sort

* Contract with projector
Local Pif = PiRaw * (d_(mu1,mu2) - q(mu1)*q(mu2)/Q2) / (d-1);
* Handle q related terms and e^2
id q.q = Q2;
id q^2 = Q2;
id q^-2 = Q2^-1;
id e^2 = 4 * pi * alpha;
.sort

* High Energy Cleanup
id Melec^2 * Q2^-1 = 0;
id ep^2 = 0;
.sort

* Done
#message ">>>"
Print +s Pif;
.sort

.end
```
